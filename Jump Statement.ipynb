```python
#1 Print numbers from 1 to 10, but stop the loop if the number is 6 using break.
for i in range(1,11):
    if i == 6:
        break
    print(i)
```

```python
#2 Loop from 1 to 20, but skip all even numbers using continue.
for i in range(1,21):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#3 Loop from 1 to 15, print only numbers divisible by 3, and stop if the number is 12 using break.
for i in range(1,16):
   if i == 12:
       break

   if i % 3 == 0:
       print(i)
```

```python
#4 Loop from 1 to 30, skip numbers that are multiples of 5 using continue.
for i in range(1,31):
    if i % 5 == 0:
        continue
    print(i, end = " ")
```

```python
#5 Create a loop from 10 to 1 and stop when the number becomes 4 using break
for i in range(10,0,-1):
   if i == 4:
       break
   print(i)
```

```python
#6 Use pass in a loop that runs from 1 to 10 to show it's a placeholder without doing anything.
for i in range(1,11):
    pass
print(i)
```

```python
#7 Use a loop to print numbers from 1 to 10, but skip printing 3 and 7 using continue.
for i in range(1,11):
    if i == 3 or i == 7:
        continue

    print(i)
```

```python
#8 Create a loop that prints numbers from 1 to 100 and uses break when the sum reaches more than 200.
sum = 0
for i in range(1,101):
    sum = sum + i
    if sum >= 200:
        break
    print(sum, end = " ")
```
