```python
#1. Write a Python program to reverse a string without using slicing.
str = "Python"
rev = ""
for ch in str:
    rev = ch + rev
print(rev)
```

```python
#2 Create a dictionary from two lists (keys & values).
keys = ["no of student","no of course"]
values = [4,3 ]
d = {}
for i in range(len(keys)):
    d[keys[i]] = values[i]
    
print(d)
```

```python
#3. Count vowels and consonants in a string(sentence) except spaces.
sentence = "Data Science"
vowels = "aeiouAEIOU"
vowel = 0
consonant = 0
for char in sentence:
    if char.isalpha():        
        if char in vowels:
            vowel += 1
        else:
            consonant += 1
print("Vowels:", vowel)
print("Consonants:", consonant)

  
```

```python
#4. Check whether a string is palindrome or not.
s = "madam"
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#5 Find the second largest element in a list without using sort().
lst = [7, 4, 3, 9, 2]
largest = second = -1
for num in lst:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num
print("Second Largest:", second)
```

```python
#6 Remove duplicate elements from a list without using set().
lst = [1, 2, 1, 3,1,2]
unique = []
for i in lst:
    if i not in unique:
        unique.append(i)
print(unique)
```

```python
#7 Separate even and odd numbers from a list
lst = [1, 2, 3, 4, 5, 6]
even = []
odd = []
for i in lst:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)
print("Even:", even)
print("Odd:", odd)
```

```python
#9 Find the index of an element in a tuple without using index().
t = (1, 2, 3, 4)
element = 3
for i in range(len(t)):
    if t[i] == element:
        print("Index:", i)
```

```python
#10 Count the frequency of elements in a tuple.
t = (1, 2, 2, 3, 3)
freq = {}
for i in t:
    freq[i] = freq.get(i, 0) + 1
print(freq)
```

```python
#13 Find the key with maximum value in a dictionary.  
d = {"a": 3, "b": 29, "c": 20}
max_key = None
max_val = -1
for k, v in d.items():
    if v > max_val:
        max_val = v
        max_key = k
print(max_key)
```

```python
#15 Given a list of integers, create a new list where each element is the product of all elements of the list except itself.
#[1, 2, 3, 4]
#[24, 12, 8, 6]

lst = [1, 2, 3, 4]
result = []
for i in range(len(lst)):
    prod = 1
    for j in range(len(lst)):
        if i != j:
            prod *= lst[j]
    result.append(prod)
print(result)
```
